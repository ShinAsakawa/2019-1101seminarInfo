```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 200) # , 500)
y1 = np.sin(x) # まずは元の信号を
y2 = y1 + np.random.randn(x.shape[0])*0.3 # ノイズを混ぜる
```

```python
v = np.ones(5)/5.0 # 移動平均をとるための配列vを設定。今回は前後５つの値を用いて平均をとる。
y3 = np.convolve(y2, v, mode='same')  # グラフを描く都合上'same'で。
#y3 = np.convolve(y2, v, mode='full')
```

```python
plt.plot(x, y1,'r',linewidth=1)
plt.plot(x, y2,'b', linewidth=1)
plt.plot(x, y3, 'y',linewidth=3)
plt.show()
```
